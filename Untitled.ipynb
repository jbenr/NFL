```python
import os
import pandas as pd

def concat_csv_files(directory_path):
    csv_list = []
    for filename in os.listdir(directory_path):
        if filename.endswith(".csv"):  # Check if the file is a CSV
            file_path = os.path.join(directory_path, filename)
            temp = pd.read_csv(file_path)
            temp['season'] = filename.split('_')[1]
            temp['week'] = filename.split('_')[2]
            csv_list.append(temp)
    
    combined_df = pd.concat(csv_list, ignore_index=True)
    
    return combined_df

# Example usage:
directory_path = "/Users/jamesreichert/PycharmProjects/NFL/data/bt/20/"
df = concat_csv_files(directory_path)
```

```python
for col in ['season','week','away_score','home_score']:
    try: sched[col] = pd.to_numeric(sched[col])
    except Exception as e: print(e)
    try: df[col] = pd.to_numeric(df[col])
    except Exception as e: print(e)

merger = pd.merge(
    df, 
    sched[['season','week','away_team','home_team','away_score','home_score']],
    on=['season','week','away_team','home_team'],
    how='left'
)

merger.drop(columns=merger.columns[0], inplace=True)

merger['actual'] = merger['home_score']-merger['away_score']
```
